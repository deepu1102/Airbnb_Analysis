```python
import pandas as pd
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.set(style="whitegrid")
```

```python

file_path = '/content/Airbnb NYC 2019.csv'
df = pd.read_csv(file_path)
```

```python
print(df)
```

```python
df.describe()
```

```python

list(df.columns)
```

```python

df.isnull().sum()
```

```python

df = df.dropna(subset=['host_name'])
```

```python
df.isnull().sum()
```

```python

df = df.dropna(subset=['name'])
```

```python
df['last_review'] = df['last_review'].astype(str).fillna('No Review')
```

```python


df['calculated_host_listings_count'] = df['calculated_host_listings_count'].fillna(df['calculated_host_listings_count'].median())
```

```python
df.isnull().sum()
```

```python
df['reviews_per_month'] = df['reviews_per_month'].fillna(df['reviews_per_month'].median())
```

```python

print(df.isnull().sum())
```

```python
df.head()
```

```python
df.tail()
```

```python
plt.figure(figsize=(6,4))
sns.countplot(data=df, x='room_type', order=df['room_type'].value_counts().index, palette='pastel')
plt.title('Room Type Distribution')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['price'], bins=100, kde=True)
plt.xlim(0, 500)  # Limit for better visualization
plt.title('Price Distribution (Capped at $500)')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.barplot(data=df, x='neighbourhood_group', y='price', estimator='mean', ci=None, palette='viridis')
plt.title('Average Price by Neighbourhood Group')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(data=df, x='room_type', y='availability_365')
plt.title('Availability per Year by Room Type')
plt.show()
```

```python
top_hosts = df['host_id'].value_counts().head(10)
plt.figure(figsize=(8,4))
sns.barplot(x=top_hosts.values, y=top_hosts.index, palette="magma")
plt.title('Top 10 Hosts by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Host ID')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.scatterplot(
    data=df.sample(10000),  # sample for speed
    x='longitude', y='latitude',
    hue='neighbourhood_group',
    alpha=0.6,
    palette='tab10'
)
plt.title('NYC Airbnb Listings by Neighbourhood Group')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Neighbourhood Group')
plt.show()
```

```python
plt.figure(figsize=(10,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Heatmap')
plt.show()
```

```python
top_neighbourhoods = df['neighbourhood'].value_counts().head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_neighbourhoods.values, y=top_neighbourhoods.index, palette='Blues_d')
plt.title('Top 10 Neighbourhoods by Number of Listings')
plt.xlabel('Number of Listings')
plt.ylabel('Neighbourhood')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.histplot(df['minimum_nights'], bins=50, kde=False)
plt.xlim(0, 30)  # Limit for better insight
plt.title('Minimum Nights Distribution (Capped at 30)')
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(data=df, x='number_of_reviews', y='price', alpha=0.3)
plt.xlim(0, 200)
plt.ylim(0, 500)
plt.title('Price vs Number of Reviews (Outliers Removed)')
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.boxplot(data=df[df['price'] < 500], x='room_type', y='price', palette='Set2')
plt.title('Price Distribution by Room Type (Under $500)')
plt.show()
```

```python
top_hosts = df['host_id'].value_counts().head(10)

plt.figure(figsize=(10,4))
sns.barplot(x=top_hosts.index.astype(str), y=top_hosts.values, palette='Oranges_r')
plt.title('Top 10 Hosts by Number of Listings')
plt.xlabel('Host ID')
plt.ylabel('Number of Listings')
plt.show()
```

```python
plt.figure(figsize=(6,4))
sns.barplot(data=df, x='room_type', y='reviews_per_month', estimator='mean', palette='PuBu')
plt.title('Average Reviews per Month by Room Type')
plt.show()
```

```python
pivot_availability = df.pivot_table(index='neighbourhood_group', columns='room_type', values='availability_365', aggfunc='mean')

plt.figure(figsize=(8,5))
sns.heatmap(pivot_availability, annot=True, fmt=".1f", cmap='YlGnBu')
plt.title('Average Availability by Room Type and Neighbourhood Group')
plt.show()
```

```python
room_type_ratio = df.groupby(['neighbourhood_group', 'room_type']).size().unstack().apply(lambda x: x / x.sum(), axis=1)

room_type_ratio.plot(kind='bar', stacked=True, figsize=(10,6), colormap='tab20')
plt.title('Room Type Ratio in Each Neighbourhood Group')
plt.ylabel('Proportion')
plt.xticks(rotation=45)
plt.legend(title='Room Type')
plt.tight_layout()
plt.show()
```

```python
top_price_neigh = df.groupby('neighbourhood')['price'].mean().sort_values(ascending=False).head(10)

plt.figure(figsize=(10,5))
sns.barplot(x=top_price_neigh.values, y=top_price_neigh.index, palette='rocket')
plt.title('Top 10 Neighbourhoods by Average Price')
plt.xlabel('Average Price')
plt.ylabel('Neighbourhood')
plt.show()
```

```python
host_df = df.groupby('host_id').agg({
    'id': 'count',
    'number_of_reviews': 'sum'
}).rename(columns={'id': 'listing_count'})

plt.figure(figsize=(8,5))
sns.scatterplot(data=host_df, x='listing_count', y='number_of_reviews', alpha=0.5)
plt.title('Host Strategy: Listings vs Total Reviews')
plt.xlabel('Listing Count')
plt.ylabel('Total Reviews')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.show()
```

📌 Conclusion     
The Airbnb NYC 2019 dataset reveals that Entire home/apt is the most popular and expensive room type, while Private rooms offer a more budget-friendly alternative. Manhattan dominates with the highest number of listings and average prices, followed by Brooklyn, which balances cost and availability. Most listings are priced between $50–$200, though a few luxury listings skew the average. A small group of top hosts controls many listings, suggesting commercial activity, yet not all receive high engagement through reviews. Listings with more reviews tend to be cheaper and more available, indicating that affordability and accessibility drive user engagement in the NYC Airbnb market.
